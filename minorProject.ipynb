```python
df.replace('manhatan','Manhattan',inplace=True)
df.replace('brookln','Brooklyn',inplace=True)
df.replace('queens','Queens',inplace=True)
df.replace('staten_island','Staten Island',inplace=True)
df.replace('bronx','Bronx',inplace=True)
```

```python
def fun(col):
  # Calculate Q1 (25th percentile) and Q3 (75th percentile)
  Q1 = col.quantile(0.25)
  Q3 = col.quantile(0.75)
  IQR = Q3 - Q1

  # Define bounds
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  return lower_bound, upper_bound
l,u=fun(df['availability 365'])
print(l,u)

# Remove outliers
df = df[(df['availability 365'] >= l) & (df['availability 365'] <= u)]
#df=df.dropna(df[df['availability]<l].index|df[df['availability]>u].index)
df
```

```python
property_types=df['room type'].value_counts()
property_types
```

```python
plt.bar(property_types.index,property_types.values)
plt.ylabel('count')
plt.xlabel('room_types')
plt.title('room_types and their count in Dataset')
plt.show()
```

```python
nh_group=df['neighbourhood group'].value_counts()
nh_group
```

```python
plt.bar(nh_group.index,nh_group.values)
plt.ylabel('count')
plt.xlabel('neighbourhood group')
plt.title('neighbourhood group and their count in Dataset')
plt.show()
```

```python
nhg_price=df.groupby('neighbourhood group')['price'].mean()
nhg_price
```

```python
plt.bar(nhg_price.index,nhg_price.values)
plt.ylabel('Avg price')
plt.xlabel('neighbourhood group')
plt.title('neighbourhood group and their price in Dataset')
for i, v in enumerate(nhg_price):
    plt.text(i, v+2 , f"{v:.2f}", ha='center')
plt.show()
```

```python
hosts=df.groupby('host name')['calculated host listings count'].sum().sort_values(ascending=False)
hosts
```

```python
plt.bar(hosts.index[:10],hosts.values[:10])
plt.ylabel('calculated host listings count')
plt.xlabel('host name')
plt.xticks(rotation=90)
plt.title('Top 10 host names and their calculated host listings count in Dataset')
```

```python
reviews=df.groupby('host_identity_verified')['review rate number'].mean()
reviews

# hosts with verified identities has more positive reviews
```

```python
plt.plot(df['price'],df['service fee'])
plt.xlabel('price')
plt.ylabel('service fee')
plt.title('price and service fee correlation')
#positive correlation
```

```python
arrn = df.groupby(['neighbourhood group', 'room type'])['review rate number'].mean()
arrn
```

```python
plt.figure(figsize=(8,10))
sns.barplot(x=arrn.index.get_level_values(0), y=arrn.values, hue=arrn.index.get_level_values(1))
plt.xlabel('Neighbourhood Group')
plt.ylabel('Average Review Rate Number')
plt.title('Average Review Rate Number by Neighbourhood Group and Room Type')
plt.legend(title='Room Type',loc=2)
plt.show()
```
